```python
dbConnection = sqlite3.connect("bookdb");
dbCursor = dbConnection.cursor()
```

```python
txtExtention = ".txt"
```

```python
tables = ["Customers", "Orderlines", "Products", "Campaigns", "Orders"]
```

```python
for table in tables:
    path = os.path.expanduser(table.lower() + txtExtention)
    file = pd.read_csv(path, delimiter="\t", encoding = 'latin1')
    file.to_sql(table, dbConnection, if_exists="replace");
```

```python
pd.read_sql_query('''SELECT o.orderdate, c.gender, c.firstname from Orders o join Customers c on o.customerid = c.customerid 
                    join Campaigns cmp on o.campaignid = cmp.campaignid ''', dbConnection)
```

```python
pd.read_sql_query('SELECT * from Products', dbConnection)
```

```python
dataFrame = pd.read_sql_query('''SELECT l.shipdate, l.billdate, l.unitprice, l.numunits, l.totalprice, 
o.orderdate, o.city from Orderlines l 
join Products p on l.productid = p.productid 
join Orders o on o.orderid = (SELECT orderid from Orders o 
join Customers c on o.customerid = c.customerid 
join Campaigns cmp on o.campaignid = cmp.campaignid)''', dbConnection)
```
